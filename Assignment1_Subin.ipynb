# Dataset Information

Given Dastaset consists of information regarding different types of Iris plants and their properties.

1. SL - Sepal length
2. SW - Sepal width
3. PL - Petal length
4. PW - Petal width

Classification of plant
    Iris Setosa
    Iris Versicolour
    Iris Virginica

# Visualizations

Scatter plot

```python
dataset ['Classification'].value_counts()
```

```python
Classification = ['Iris-versicolor','Iris-virginica',
'Iris-setosa']
```

```python
for i in range(3):
    x = dataset[dataset['Classification'] == Classification[i]]
    plt.scatter(x['SL'], x['SW'], label = Classification[i])
    plt.xlabel('SL')
    plt.ylabel('SW')
    plt.legend()
```

```python
for i in range(3):
    x = dataset[dataset['Classification'] == Classification[i]]
    plt.scatter(x['PL'], x['PW'], label = Classification[i])
    plt.xlabel('PL')
    plt.ylabel('PW')
    plt.legend()
```
